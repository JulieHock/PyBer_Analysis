```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
#Set x axis to a list of strings for each month
x_axis=["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis=[10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax=plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax=plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create teh plot and add a label for the legend
plt.plot(x_axis, y_axis, label="Boston")

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create a vertical bar chart
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label="boston")
#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create a title
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()
```

```python
# create the plot
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Create Pie Chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Format Pie Chart
#explode = pull a piece of the pie apart from the others
plt.subplots(figsize=(8,8))
explode_values=(0,0,0,0,0,0,.2,0,0,0,0,0)
#autopct displays % to 1 decimal point
plt.pie(y_axis, explode=explode_values,labels=x_axis, autopct="%.1f%%")
plt.show()
```

```python
# Assign 12 colors, one for each month
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values=(0,0,0,0,0,0,.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

plt.show()
```
